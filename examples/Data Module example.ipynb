```python
def print_batch(images, labels=None):
    fig, axes = plt.subplots(3, 3, figsize=(15, 15))
    
    for ax, image in zip(axes.flatten(), images):
        ax.set_xticks([])
        ax.set_yticks([])
        ax.imshow(image)
        if labels is not None:
            ax.set_title("Class: {}".format(label_to_name_map[label]))
```
